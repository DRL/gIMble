```python
import itertools
```

```python
def powerset(s):
    x = len(s)
    myList = [];
    for i in range(1 << x):
        myList.append( [s[j] for j in range(x) if (i & (1 << j))])
    return myList
```

```python
branches = powerset(['a','b','c','d','e','f','g'])
#branches = [ b for b in powerset(['a','b','c','d'])]
branches[1:]
```

```python
myDic = {}
for b in branches:
    myDic[frozenset(b)] = var('z_'+''.join('%s'%s for s in b))
myDic[frozenset(['a', 'b', 'c', 'd'])]
```

```python
def lap_exp(branchList):
    # here branchList is the list of dictionary entries which refer to a symbolic variable in sage
    mySum = sum([myDic[frozenset(b)] for b in branchList])
    myLambda = binomial(len(branchList),2)
    return 1/(myLambda + mySum)
```

```python
lap_exp([myDic[frozenset({'d','c','a'})],var('y'),var('z')])
```

```python
def coalesce(branchList,coalList):
    nonCoal = [b for b in branchList if b not in coalList]
    coal = [coalList[0]+coalList[1]]
    return nonCoal + coal
    
def makeGenFunction(branchList):
    #branch list is the list of current branch types, e.g. [['a'],['b','c'],[d]] is an
    # 'a bc d' configuration of the current genealogical history
    coalEvents = list(itertools.combinations(branchList,2))
    ancestors = [ coalesce(branchList,coalPair) for coalPair in coalEvents]
    ## the ancestors are what is generated by makecoalEvents 
    # in my initial implementation of the GF method in mathematica
    
    if (len(branchList) == 2):
        return lap_exp(branchList)
    else:
        phi = lap_exp(branchList)*sum( [ makeGenFunction(ancestorList) for ancestorList in ancestors])
        return phi
        
    return ancestors
```
